```python
max_repetitions = df['Watched Date'].value_counts().max()
most_frequent_item = df['Watched Date'].value_counts().idxmax()
print(f"On the day '{most_frequent_item}' i watched {max_repetitions} films the most ever since i started using letterboxd ")
```

```python
rewatched_movie_names = df[df['Rewatch'].isin(['Yes', True, 'true'])]['Name'].unique()
rewatch_counts = df[df['Name'].isin(rewatched_movie_names)]['Name'].value_counts()
rewatch_summary = rewatch_counts.reset_index()
rewatch_summary.columns = ['Movie Title', 'Total Times Rewatched']
rewatch_summary.head(20)
```
